```python
def my_function(*numbers): 
    if len(numbers) == 0:
        return None
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num
print(my_function(3, 7, 2, 9, 1))
```

```python
def my_function(username, **details):
    print("Username:", username)
    print("Additional details:")
    for key, value in details.items():
        print(" ", key + ":", value)
my_function("emil123", age = 25, city = "Oslo", hobby = "coding")
```

```python
def changecase(func):
    def myinner():
        return func().upper()
    return myinner
@changecase
def myfunction():
    return "Hello Sally"
@changecase
def otherfunction():
    return "I am speed!"
print(myfunction())
print(otherfunction())
```
