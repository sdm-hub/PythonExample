```python
import re
txt = "The rain in Spain"
x = re.search("\s", txt)
print("The first white-space character is located in position:", x.start())
```

```python
import re
txt = "The rain in Spain"
x = re.split("\s", txt, 1)
print(x)
```

```python
import re
txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span())
```

```python
import re
txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.group())
```
